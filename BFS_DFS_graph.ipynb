```python

class Graph:
    
    graph_dict={}
    
    def addEdge(self,node,neighbour):  
        if node not in self.graph_dict:
            self.graph_dict[node]=[neighbour]
        else:
            self.graph_dict[node].append(neighbour)
            
        print(self.graph_dict)   
            

  
    def BFS(self,s):
        
        visited={}
        for i in self.graph_dict:
            visited[i]=False
        queue=[]
        queue.append(s)
        visited[s]=True
        while len(queue)!=0:
            s=queue.pop(0)
            for node in self.graph_dict[s]:
                if visited[node]!=True:
                    visited[node]=True
                    queue.append(node)
            print(s,end=" ")
                
       
    def DFS(self,s):
            visited={}
            for i in self.graph_dict:
                visited[i]=False
            queue=[]
            queue.append(s)
            visited[s]=True
            while len(queue)!=0:
                s=queue.pop(0)
                for node in self.graph_dict[s]:
                    for i in self.graph_dict[node]:
                        if visited[i]==False:
                            visited[i]=True
                       
                            queue.append(i)
                print("DFS")         
                print(s,end=" ")  
                print("\n")

                
                
g= Graph()
g.addEdge('1', '2')
g.addEdge('1', '3')
g.addEdge('2', '3')
g.addEdge('2', '1')
g.addEdge('3', '1')
g.addEdge('3', '2')
g.addEdge('3', '4')
g.addEdge('4', '3')
g.BFS('1')
g.DFS('1')
```
